```python
category=df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']]
category
```

# Data preprocess

```python
onehotencoder=OneHotEncoder()
category=onehotencoder.fit_transform(category).toarray()
category
```

```python
category=pd.DataFrame(category)
```

```python
category
```

```python
 numeric = df.select_dtypes(include= 'int64')
 numeric.drop(['Attrition'], axis=1,inplace=True)
```

```python
numeric
```

```python
X_all=pd.concat([category,numeric],axis=1)
```

```python
scaler=MinMaxScaler()
X=scaler.fit_transform(X_all)
```

```python
X
```

```python
y=df['Attrition']
y
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.23)
```

```python
X_train.shape, X_test.shape, y_train.shape,y_test.shape
```

# Modelling

## Logistic regression

```python
logistic=LogisticRegression()
logistic.fit(X_train,y_train,)
```

```python
y_pred=logistic.predict(X_test)
y_pred
```

```python
y_test
```

```python
accuracy_score(y_test, y_pred)
```

```python
# Making the confusion matrix
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
# precision
precision_score(y_test, y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
f1_score(y_test,y_pred, average='macro')
```

```python
print(classification_report(y_test,y_pred))
```

## Random forest

```python
forest=RandomForestClassifier()
forest.fit(X_train,y_train)
```

```python
y_pred=forest.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

## Artificial neural network

```python
rede_neural=tf.keras.models.Sequential()
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu', input_shape=(50,)))
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu'))
rede_neural.add(tf.keras.layers.Dense(units=25, activation='relu'))
rede_neural.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
rede_neural.fit(X_train,y_train,epochs=200)
```

```python
y_pred=rede_neural.predict(X_test)
```

```python
y_pred=(y_pred>=0.5)
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot=True);
```

```python
print(classification_report(y_test,y_pred))
```
